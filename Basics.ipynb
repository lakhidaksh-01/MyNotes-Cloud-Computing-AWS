# Cloud Computing Explained

Instead of buying a powerful computer, storage, servers, and maintaining everything yourself…
you just rent it from the internet whenever you need it.

That’s cloud computing.

👉 In simple words:
You are using someone else’s computer (server) over the internet

Example you already use daily:

- Google Drive → storage
- Netflix → streaming
- Gmail → email

All of this runs on cloud.

## Definition

Cloud computing = using computer services from the internet instead of your own computer. It allows users to access and use shared resources, such as servers, storage, databases, networking, software, and analytics, without having to manage the underlying infrastructure.

## Traditional vs. Cloud Computing

| Feature       | Traditional 🏢 | Cloud ☁️       |
| ------------- | -------------- | -------------- |
| Setup         | Buy everything | Rent           |
| Maintenance   | You            | Provider       |
| Cost          | High upfront   | Pay-as-you-use |
| Scalability   | Difficult      | Easy           |
| Flexibility   | Low            | High           |

## A Simple Way to Differentiate

A simple way to differentiate between cloud and On-Premises is that the software uses internet or not.

- If **not** uses internet then it is **on-premises**  
  *Example: Microsoft Word offline version, calculator on phone.*

- If **uses** internet then it is **cloud computing**  
  *Example: Microsoft Word Online (Word 365), Instagram, Facebook.*

## ⭐ Characteristics (VERY IMPORTANT)

👉 Shortcut: **O M R R M**

1.  **On-demand self-service**  
    *Example:* You want a server → click → boom, it’s ready.

2.  **Measured service (pay-as-you-use)**  
    You only pay for what you use.

3.  **Resource pooling (share resources among multiple users)**  
    Shared resources are allocated dynamically to meet the needs of multiple users.

4.  **Rapid elasticity (scale up/down)**  
    You can quickly scale your resources up or down based on your needs.

5.  **Multi-tenancy**  
    Multiple users (tenants) can share the same infrastructure while keeping their data and applications separate and secure.

**Note:** Resource pooling provides shared resources, while multi-tenancy allows multiple users to use those shared resources securely.

## Cloud Models

2 Types of cloud models:

### 1. Deployment Models (where it is located) : 4 types further

| Model     | Meaning     | Example       |
| --------- | ----------- | ------------- |
| Public    | Open to all | AWS           |
| Private   | Single org  | Bank server   |
| Hybrid    | Mix of both | Company setup |
| Community | Shared orgs | Govt          |

**Hybrid cloud** = combination of public + private cloud

*Example:*
Company stores sensitive data in private cloud
Uses public cloud for normal work

### 2. Service Models (what services it provides)

| Short Form | Full Form                   | Easy Real-Life Examples                                                                    |
| ---------- | --------------------------- | -------------------------------------------------------------------------------------------|
| IaaS       | Infrastructure as a Service | Renting a computer/server online, like running your own website or game server, or aws ec2 |
| PaaS       | Platform as a Service       | Coding apps online without setup (like coding projects on online platforms)                |
| SaaS       | Software as a Service       | Gmail, Google Docs, YouTube                                                                |

## Virtualization

Virtualization is the process of running multiple operating systems on a single physical machine using a hypervisor.

### Type 1 vs Type 2 Hypervisor (VERY IMPORTANT)

| Feature | Type 1 (Bare Metal)      | Type 2 (Hosted)      |
| ------- | ------------------------ | -------------------- |
| Runs On | Directly on hardware     | On top of OS         |
| Speed   | Fast                     | Slower               |
| Use     | Companies / Data centers | Personal use         |
| Example | VMware ESXi              | Oracle VM VirtualBox |

### Benefits of Virtualization

- Cost saving (less hardware needed) 💰
- Better resource utilization 📊
- Run multiple OS on one machine 💻
- Easy testing & development 🧪
- Scalability 📈
